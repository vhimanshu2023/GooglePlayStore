# **Deal with missing value and some other values**

```
df.isnull().sum()
```

```
ndf = df.isnull()
sns.heatmap(ndf)
plt.show()
```

```
ndf1 = df.notnull()
sns.heatmap(ndf1)
plt.show()
```

```
df.fillna(
    {
    'Rating' : 0.0,
    'Type' : df['Type'].fillna(method='ffill'),
    'Content Rating' : df['Content Rating'].fillna(method='ffill'),
    'Current Ver' : df['Current Ver'].fillna(method='ffill'),
    'Android Ver' : df['Android Ver'].fillna(method='ffill')
    }
)
```

```
df['Size'] = df['Size'].str.replace('M','')
df['Installs'] = df['Installs'].str.replace('+','')
df['Installs'] = df['Installs'].str.replace(',','')
```

```
df['Rating'].sort_values(ascending=False)
df.drop([10472],inplace=True)
```

 Total Number of App Titles Contain Astrology

```
len(df.loc[df['App'].str.contains('Astrology')])
```

Total Number of Unique Category

```
len(df['Category'].unique())
```

Highest Avg. Rating

```
df.groupby('Category')['Rating'].mean().max()
```

Total Number of App having 5 Star Rating

```
len(df.loc[df['Rating'] == 5])
```

 Average Value of Reviews

```
df['Reviews'] = df['Reviews'].str.replace('M','')
```

```
df['Reviews'] = df['Reviews'].astype(float)
```

```
df['Reviews'].dtype
```

```
df['Reviews'].mean()
```

```
df.groupby('Type')['Type'].count()
```

# Visualization

```
sns.countplot(data=df ,x='Rating')
plt.show()
```

```
sns.countplot(data=df ,x='Rating',hue='Type')
plt.show()
```

```
sns.pairplot(data=df,hue='Type')
plt.show()
```

```
cnt = df.groupby('Category').count().reset_index()
sns.relplot(x=cnt['Category'],y=cnt['App'],data=cnt,hue='Type')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
